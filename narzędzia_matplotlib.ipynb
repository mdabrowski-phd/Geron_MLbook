# Rysowanie pierwszego grafu

Najpierw upewnijmy się, że notatnik ten będzie działał w środowiskach Python 2 i 3:

```python
from __future__ import division, print_function, unicode_literals
```

Przede wszytkim musimy importować bibliotekę `matplotlib`.

```python
import matplotlib
```

Moduł matplotlib może wyświetlać wykresy za pomocą wielu wbudowanych bibliotek graficznych, takich jak Tk, wxPython itd. Jeżeli obsługujesz środowisko Python z poziomu wiersza poleceń, wykresy będą zazwyczaj wyświetlane w oddzielnym oknie. W przypadku notatników Jupyter mogą być one wyświetlane w samym notatniku, jeśli wprowadzimy magiczną komendę `%matplotlib inline`.

```python
%matplotlib inline
# matplotlib.use("TKAgg")  # użyj tego wiersza w programie, jeśli wolisz generować wykresy za pomocą silnika Tk.
```

Czas na nasz pierwszy wykres! :)

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 4, 9, 5, 3])
plt.show()
```

Tak, wystarczy podać funkcję `plot` z jakimiś danymi, a następnie wywołać funkcję `show`!

Jeśli podamy funkcji `plot` tablicę z danymi, zostaną one wykorzystane jako współrzędne na osi rzędnych, natomiast za współrzędne w osi odciętych będą służyć indeksy poszczególnych punktów danych.
Możemy również przekazywać dwie tablice: jedna będzie zawierać współrzędne na osi odciętych `x`, a druga - na osi rzędnych `y`:

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.show()
```

Osie są automatycznie dopasowywane do rozmiaru danych. Chcielibyśmy zrobić trochę dodatkowego miejscda na wykresie, dlatego wywołajmy funkcję `axis` w celu zmiany zakresu każdej osi `[xmin, xmax, ymin, ymax]`.

```python
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.axis([-4, 6, 0, 7])
plt.show()
```

Stwórzmy teraz wykres funkcji matematycznej. Wykorzystamy funkcję `linspace` stanowiącą część modułu NumPy do stworzenia tablicy `x` zawierającej 500 wartości zmiennoprzecinkowych w zakresie od -2 do 2, a następnie stworzymy drugą tablicę, `y`, której wartości będą stanowić kwadrat wartości `x` (więcej informacji na temat modułu NumPy znajdziesz w [samouczku NumPy](narzędzia_numpy.ipynb)).

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

Trochę smutny ten wykres; dodajmy tytuł, etykiety osi x i y, a także siatkę.

```python
plt.plot(x, y)
plt.title("Funkcja kwadratowa")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

# Styl i kolor linii

Domyślnie moduł matplotlib rysuje linie pomiędzy następującymi po sobie punktami.

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0])
plt.axis([-10, 110, -10, 140])
plt.show()
```

Za pomocą trzeciego argumentu możemy zmienić styl i kolor linii. Na przykład `"g--"` oznacza "zieloną kreskowaną linię".

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

W bardzo prosty sposób możemy rysować wiele linii na jednym wykresie: wystarczy przekazać argumenty w postaci `x1, y1, [styl1], x2, y2, [styl2], ...`

Przykład:

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-", [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

Albo po prostu wywoływać funkcję `plot` wiele razy przed wywołaniem funkcji `show`.

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

Jesteśmy w stanie również rysować zwykłe punkty zamiast linii. Poniżej prezentuję przykład zawierający zielone kreski, czerwone kropki i niebieskie trójkąciki.
Pełną listę ustawień stylów i kolorów znajdziesz w [dokumentacji](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot).

```python
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
plt.show()
```

Funkcja `plot` zwraca listę obiektów typu `Line2D` (po jednym na każdą linię). Możemy wyznaczać każdej z tych linii dodatkowe atrybuty, takie jak grubość linii, styl kreskowania czy przezroczystość (współczynnik alfa). Również pełną listę atrybutów znajdziesz w [dokumentacji](http://matplotlib.org/users/pyplot_tutorial.html#controlling-line-properties).

```python
x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

# Wykresy podrzędne
Wykres matplotlib może przechowywać wiele wykresów podrzędnych. Są one zorganizowane w postaci siatki. Aby stworzyć wykres podrzędny, wystarczy wywołać funkcję `subplot` i określić liczbę rzędów i kolumn rysunku wynikowego, a także indeks wykresu podrzędnego (począwszy od 1, następnie od lewej do prawej oraz od góry w dół). Zwróć uwagę, że moduł pyplot śledzi obecnie aktywny węzeł podrzędny (możesz go sprawdzić za pomoć metody `plt.gca()`), zatem w chwili wywołania funkcji `plot` zostanie wygenerowany *aktywny* wykres podrzędny.


```python
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2, 2, 1)  # Dwa rzędy, dwie kolumny, pierwszy wykres podrzędny = górny lewy
plt.plot(x, x)
plt.subplot(2, 2, 2)  # Dwa rzędy, dwie kolumny, drugi wykres podrzędny = górny prawy
plt.plot(x, x**2)
plt.subplot(2, 2, 3)  # Dwa rzędy, dwie kolumny, trzeci wykres podrzędny = dolny lewy
plt.plot(x, x**3)
plt.subplot(2, 2, 4)  # Dwa rzędy, dwie kolumny, czwarty wykres podrzędny = dolny prawy
plt.plot(x, x**4)
plt.show()
```

* Zwróć uwagę, że zapis `subplot(223)` stanowi skrót funkcji `subplot(2, 2, 3)`.

Łatwo tworzyć wykresy podrzędne rozszerzane na wiele komórek sieci, na przykład:

```python
plt.subplot(2, 2, 1)  # Dwa rzędy, dwie kolumny, pierwszy wykres podrzędny = górny lewy
plt.plot(x, x)
plt.subplot(2, 2, 2)  # Dwa rzędy, dwie kolumny, drugi wykres podrzędny = górny prawy
plt.plot(x, x**2)
plt.subplot(2, 1, 2)  # Dwa rzędy, *jedna* kolumna, drugi wykres podrzędny = dolny
plt.plot(x, x**3)
plt.show()
```

Gdybyśmy potrzebowali bardziej zaawansowanych metod pozycjonowania wykresów podrzędnych, moglibyśmy skorzystać z funkcji `subplot2grid` zamiast funkcji `subplot`. Określamy liczbę rzędów i kolumn w siatce, następnie pozycję wykresu podrzędnego w tej siatce (górny lewy = (0,0)), oraz, ewentualnie, liczbę komórek siatki, na jaką ma się ten wykres rozszerzać. Przykładowo:

```python
plt.subplot2grid((3,3), (0, 0), rowspan=2, colspan=2)
plt.plot(x, x**2)
plt.subplot2grid((3,3), (0, 2))
plt.plot(x, x**3)
plt.subplot2grid((3,3), (1, 2), rowspan=2)
plt.plot(x, x**4)
plt.subplot2grid((3,3), (2, 0), colspan=2)
plt.plot(x, x**5)
plt.show()
```

Jeśli potrzebujesz dodatkowych możliwości w pozycjonowaniu wykresów podrzędnych, zajrzyj do [dokumentacji GridSpec](http://matplotlib.org/users/gridspec.html)

# Wiele wykresów naraz
Możliwe jest również generowanie wielu wykresów naraz. Każdy z nich może zawierać co najmniej jeden wykres podrzędny. Moduł matplotlib w domyśle tworzy obiekt `figure(1)` automatycznie. Podczas przełączania się pomiędzy wykresami, moduł pyplot śledzi obecnie aktywnie wykres (możemy go sprawdzić wywołując metodę `plt.gcf()`), i aktywny wykres podrzędny staje się wykresem bieżącym.

```python
x = np.linspace(-1.4, 1.4, 30)

plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title("Funkcje kwadratowa i sześcienna")
plt.subplot(212)
plt.plot(x, x**3)

plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x, x**5)
plt.title("y = x**5")

plt.figure(1)      # wracamy do pierwszego rysunku, jego bieżącym wykresem podrzędnym jest 212 (dolny)
plt.plot(x, -x**3, "r:")

plt.show()
```

Na szczęście moduł Pyplot pozwala całkowicie ignorować maszynę stanów, dzięki czemu jesteśmy w stanie pisać cudownie jawny kod. Wystarczy wywołać funkcję `subplots`, a następnie użyć zwracanych obiektu wykresu i listy obiektów osi. Więcej magii nam nie trzeba! Na przykład:

Z kolei moduł pylab pozwala na importowanie modułów matplotlib.pyplot i NumPy do tej samej przestrzeni nazw. Znajdziesz wiele przykładów korzystania z modułu pylab, mimo że nie jest to już zalecane (ponieważ *wyrażone wprost* importowanie klas jest lepsze od *domniemanego*).

# Rysowanie tekstu
Możesz wywołać funkcję `text` w celu umieszczania teksti w dowolnym miejscu wykresu. Określ tylko jego współrzędne oraz, ewentualnie, dodatkowe atrybuty. Dowolny tekst w module matplotlib może zawiera równania TeX (więcej informacji znajdziesz w [dokumentacji](http://matplotlib.org/users/mathtext.html).

```python
x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2

plt.plot(x, x**2, "b-", px, py, "ro")

plt.text(0, 1.5, "Funkcja kwadratowa\n$y = x^2$", fontsize=20, color='blue', horizontalalignment="center")
plt.text(px - 0.08, py, "Piękny punkt", ha="right", weight="heavy")
plt.text(px, py, "x = %0.2f\ny = %0.2f"%(px, py), rotation=50, color='gray')

plt.show()
```

* Uwaga: wyrażenie `ha` jest aliasem atrybutu `horizontalalignment`

Pozostałe właściwości tekstu zostały opisane w [dokumentacji](http://matplotlib.org/users/text_props.html#text-properties).

Zaznaczanie elementów tekstowych do wykresu, takich jak powyższy piękny punkt, jest powszechnie stosowanym procederem. Funkcja `annotate` znacznie nam to ułatwia: wskaż jedynie interesujące Cię miejsce oraz położenie tekstu, a także, dodatkowo, atrybuty tekstu i strzałki.

```python
plt.plot(x, x**2, px, py, "ro")
plt.annotate("Piękny punkt", xy=(px, py), xytext=(px-1.3,py+0.5),
                           color="green", weight="heavy", fontsize=14,
                           arrowprops={"facecolor": "lightgreen"})
plt.show()
```

Możesz również wstawić tekst w prostokąt ograniczający za pomocą atrybutu `bbox`:

```python
plt.plot(x, x**2, px, py, "ro")

bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
plt.text(px-0.2, py, "Piękny punkt", bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black", fc="#EEEEFF", lw=5)
plt.text(0, 1.5, "Funkcja kwadratowa\n$y = x^2$", fontsize=20, color='black', ha="center", bbox=bbox_props)

plt.show()
```

Możesz dla zabawy pobawić się stylem [xkcd](http://xkcd.com), definiując wykres w sekcji `with plt.xkcd()`:

```python
with plt.xkcd():
    plt.plot(x, x**2, px, py, "ro")

    bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
    plt.text(px-0.2, py, "Piękny punkt", bbox=bbox_props, ha="right")

    bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black", fc="#EEEEFF", lw=5)
    plt.text(0, 1.5, "F. kwadratowa\n$y = x^2$", fontsize=20, color='black', ha="center", bbox=bbox_props)

    plt.show()
```

# Legendy
Najprostszym sposobem dodania legendy jest wyznaczenie etykiety dla wszystkich linii, a następnie wywołanie funkcji `legend`.

```python
x = np.linspace(-1.4, 1.4, 50)
plt.plot(x, x**2, "r--", label="Funkcja kwadratowa")
plt.plot(x, x**3, "g-", label="Funkcja sześcienna")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

# Skale nieliniowe
Obsługiwane są tutaj również skale nieliniowe, takie jak logarytmiczna, czy logitowa.

```python
x = np.linspace(0.1, 15, 500)
y = x**3/np.exp(2*x)

plt.figure(1)
plt.plot(x, y)
plt.yscale('linear')
plt.title('liniowa')
plt.grid(True)

plt.figure(2)
plt.plot(x, y)
plt.yscale('log')
plt.title('logarytmiczna')
plt.grid(True)

plt.figure(3)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logitowa')
plt.grid(True)

plt.figure(4)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.05)
plt.title('symlog')
plt.grid(True)

plt.show()
```

# Jednostki na osiach
Osie mają niewielkie znaczniki odległości zwane jednostkami. Domyślne ustawienia zazwyczaj w zupełności wystarczą, gdyż umieszczanych jest od ~5 do 8 jednostek rozmieszczonych w rozsądnej odległości od siebie.

Czasami jednak potrzebujemy większej kontroli (np. na powyższym wykresie w skali logitowej widnieje zbyt wiele jednostek). Na szczęście mamy nad nimi pełną władzę. Możemy nawet włączyć jednostki pośrednie.



```python
x = np.linspace(-2, 2, 100)

plt.figure(1, figsize=(15,10))
plt.subplot(131)
plt.plot(x, x**3)
plt.grid(True)
plt.title("Domyślne jednostki")

ax = plt.subplot(132)
plt.plot(x, x**3)
ax.xaxis.set_ticks(np.arange(-2, 2, 1))
plt.grid(True)
plt.title("Ręcznie definiowane jednostki na osi x")

ax = plt.subplot(133)
plt.plot(x, x**3)
plt.minorticks_on()
ax.tick_params(axis='x', which='minor', bottom='off')
ax.xaxis.set_ticks([-2, 0, 1, 2])
ax.yaxis.set_ticks(np.arange(-5, 5, 1))
ax.yaxis.set_ticklabels(["min", -4, -3, -2, -1, 0, 1, 2, 3, "max"])
plt.title("Ręcznie definiowane jednostki i ich podpisy\n(plus jednostki pośrednie) na osi y")


plt.grid(True)

plt.show()
```

# Rzutowanie biegunowe
Rzutowanie biegunowe polega jedynie na wyznaczeniu wartości `"polar"` dla atrybutu `projection` podczas tworzenia wykresu podrzędnego.

```python
radius = 1
theta = np.linspace(0, 2*np.pi*radius, 1000)

plt.subplot(111, projection='polar')
plt.plot(theta, np.sin(5*theta), "g-")
plt.plot(theta, 0.5*np.cos(20*theta), "b-")
plt.show()
```

# Rzutowanie trójwymiarowe

# Linie
Jak już wiemy, możemy rysować linie tak jak do tej pory, za pomocą funkcji `plot`. Często jednak wygodniej jest stworzyć funkcję rysującą nieskończoną (pozornie) linię przez cały układ współrzędnych, podając nachylenie i punkt przecięcia. Dostępne są również funkcje `hlines` i `vlines` generujące, odpowiednio, poziome i pionowe odcinki. Na przykład:

```python
from numpy.random import randn

def plot_line(axis, slope, intercept, **kargs):
    xmin, xmax = axis.get_xlim()
    plt.plot([xmin, xmax], [xmin*slope+intercept, xmax*slope+intercept], **kargs)

x = randn(1000)
y = 0.5*x + 5 + randn(1000)*2
plt.axis([-2.5, 2.5, -5, 15])
plt.scatter(x, y, alpha=0.2)
plt.plot(1, 0, "ro")
plt.vlines(1, -5, 0, color="red")
plt.hlines(0, -2.5, 1, color="red")
plot_line(axis=plt.gca(), slope=0.5, intercept=5, color="magenta")
plt.grid(True)
plt.show()
```

# Animacje
Moduł matplotlib jest używany głównie do generowania wykresów, możemy również jednak za jego pomocą wyświetlać animacje, w zależności od używanej biblioteki graficznej. W przypadku notatnika Jupyter musimy skorzystać z biblioteki `nbagg` w celu korzystania z interaktywnych funkcji modułu matplotlib, w tym również z animacji. Musimy także importować klasę `matplotlib.animation`.

```python
%matplotlib nbagg
import matplotlib.animation as animation
```

W tym przykładzie najpierw bierzemy się za tworzenie punktów danych, następnie przygotowujemy pusty wykres, definiujemy funkcję aktualizacji, która będzie wywoływana przy każdym przebiegu animacji, a na końcu dodajemy do wykresu animację poprzez utworzenie konstruktora `FuncAnimation`.

Konstruktor `FuncAnimation` pobiera wykres, funkcję aktualizacji i dodatkowe argumenty. Określamy tu, że chcemy stuklatkową animację, w której odstępy pomiędzy poszczególnymi klatkami mają wynosić 20 ms. Po każdym przebiegu konstruktor `FuncAnimation` wywołuje naszą funkcję aktualizacji i przekazuje jej numer klatki `num` (w naszym przypadku od 0 do 99), po czym wstawiamy dodatkowe argumenty za pomocą zmiennej `fargs`.

Nasza funkcja aktualizacji wyznacza jedynie dane linii jako pierwsze punkty danych `num` (zatem dane są stopniowo rysowane), a z nudów wstawimy także niewielką liczbę losową do każdego punktu danych, dzięki czemu linia będzie nieco "szaleć".

```python
x = np.linspace(-1, 1, 100)
y = np.sin(x**2*25)
data = np.array([x, y])

fig = plt.figure()
line, = plt.plot([], [], "r-") # zaczynamy od pustego wykresu
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.plot([-0.5, 0.5], [0, 0], "b-", [0, 0], [-0.5, 0.5], "b-", 0, 0, "ro")
plt.grid(True)
plt.title("Zdumiewająca animacja")

# ta funkcja będzie wywoływana przy każdej iteracji
def update_line(num, data, line):
    line.set_data(data[..., :num] + np.random.rand(2, num) / 25)  # tworzymy jedynie wykres pierwszych punktów danych `num`
    return line

line_ani = animation.FuncAnimation(fig, update_line, frames=100, fargs=(data, line), interval=67)
plt.show()
```

# Zapisywanie animacji do plików wideo
Moduł matplotlib wykorzystuje biblioteki firm trzecich (np.  [FFMPEG](https://www.ffmpeg.org/)) lub klasę `mencoder` do zapisywania plików wideo. W tym przypadku skorzystamy z biblioteki FFMPEG, dlatego należy ją najpierw zainstalować.

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)
line_ani.save('my_wiggly_animation.mp4', writer=writer)
```
